```python
label1 = merge01[ merge01["label_x"] == merge01["label_y"] ]
label2 = merge01[ (merge01["label_x"].isna()) | (merge01["label_y"].isna())  ]
labelled = pd.merge(label1, label2, how = "outer")
labelled.head()
```

```python
def labelling(row):
    if (row.label_x == 1) | (row.label_x == 0):
        return row.label_x
    else:
        return row.label_y
    
def labelling1(row):
    return row["label_A"]


labelled["label"] = labelled.apply(labelling, axis =1)
labelled.drop(["label_x", "label_y"], axis = 1, inplace=True)
```
